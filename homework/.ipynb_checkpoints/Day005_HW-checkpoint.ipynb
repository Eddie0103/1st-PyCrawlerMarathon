* 撰寫程式存取 API 且解析 JSON 格式資料


## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
#coding: utf-8
import requests
import json

url = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
res = url.text

json = json.loads(res)
print('這個 API 一次會回傳幾筆資料？\n%s' % len(json))
print('每一筆資料包含哪些欄位？\n%s' % len(json[0]))
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for i in json:
    print('標題：{0}, 貼文時間：{1}, 留言人數：{2}, 按讚人數：{3}'.format(i['title'], i['createdAt'], i['commentCount'], i['likeCount']))
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
com_count = 0
like_count = 0

for i in json:
    com_count += i['commentCount']
    like_count += i['likeCount']

print('平均留言人數：{0}\n平均按讚人數：{1}'.format(com_count, like_count))
```
