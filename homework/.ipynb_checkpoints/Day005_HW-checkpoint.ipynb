* è«‹åˆ©ç”¨ API: https://www.dcard.tw/_api/forums/pet/posts?popular=true å›ç­”ä¸‹åˆ—å•é¡Œï¼š

1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

```python
#coding: utf-8
import requests
import json

url = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
res = url.text

hot = json.loads(res)

print('é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿ\n%s' % len(hot))
print('æ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ\n%s' % (hot[0].keys()))
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for i in hot:
    print('æ¨™é¡Œï¼š{0}\nè²¼æ–‡æ™‚é–“ï¼š{1}\nç•™è¨€äººæ•¸ï¼š{2}\næŒ‰è®šäººæ•¸ï¼š{3}\n'.format(i['title'], i['createdAt'], i['commentCount'], i['likeCount']))
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
com_count = 0
like_count = 0

for i in hot:
    com_count += i['commentCount']
    like_count += i['likeCount']

print('ç†±é–€æ–‡ç« ï¼š\nå¹³å‡ç•™è¨€äººæ•¸ï¼š{0}\nå¹³å‡æŒ‰è®šäººæ•¸ï¼š{1}'.format((com_count/len(hot)), (like_count/len(hot))))


url_off = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
res_off = url_off.text

cold = json.loads(res_off)

com_count_off = 0
like_count_off = 0

for i in cold:
    com_count_off += i['commentCount']
    like_count_off += i['likeCount']

print('\néç†±é–€æ–‡ç« ï¼š\nå¹³å‡ç•™è¨€äººæ•¸ï¼š{0}\nå¹³å‡æŒ‰è®šäººæ•¸ï¼š{1}'.format((com_count_off/len(cold)), (like_count_off/len(cold))))
```
